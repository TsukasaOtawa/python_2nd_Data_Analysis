## 結合する

````python
# ファイルからの読み込み
import pandas as pd
df = pd.read_csv("../testSJIS.csv", encoding="Shift_JIS", index_col=0)
df
````

````python
# 結合（補完 : merge）
dfKS = df[["国語","数学"]]
dfE    = df[["英語"]]
dfALL = pd.merge(dfKS, dfE, on="名前", how="left")
dfALL
# dfALL = pd.concat([dfKS, dfE], axis=1) # concat axis=1 にしても merge と同等の挙動になる
# dfALL
````

````python
# 結合(追加 : concat)
dfALL = pd.concat([dfKS, dfKS])
dfALL
````

## Group BY

````python
# groupby
dfALL.groupby(["名前"]).median()[["国語", "数学"]]
````

````python
# groupby (Index を引き継がない)
dfALL.groupby(["名前"], as_index=False).median()[["国語", "数学"]]
````
